```python
#importing libraries
import numpy as np
import pandas as pd
import random
import math
np.random.seed(42)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#declaring various parameters of self organizing map
x_len=100#number of neurons in x direction
y_len=100#number of neurons in y direction
n_color=24#number of colors
t_epoch=1000#total number of epochs
a_0=0.8#initial learning rate
feat_size=3# feature size
```

```python
#dist function calculates the distance between arrays
def dist(x,y):
    val=np.linalg.norm(x-y)
    return val
```

```python
# alpha_update function implements the adapting learning rate
def alpha_update(k):
    return a_0*np.exp(-(k-1)/1000)
```

```python
# sigma_update function updates radius of neighbourhood function
def sigma_update(k,sig_0):
    return sig_0*np.exp(-(k-1)/1000)
```

```python
# wt_update function updates weight 
def wt_update(m,n,epoch_num,sig,ksom,sample_inp):
    sig_val=sigma_update(epoch_num,sig)
    for x in range(x_len):
        for y in range(y_len):
            #finding the points within the radius of neighbourhood function
            if dist(np.array([x,y]),np.array([m,n]))<sig_val:
                neigh_dist=dist(np.array([x,y]),np.array([m,n]))
                #calculating the neighbourhood funtion
                neigh_inf=np.exp(-(neigh_dist*neigh_dist)/(2*sig_val*sig_val))
                ksom[x][y]=ksom[x][y]+alpha_update(epoch_num)*neigh_inf*(sample_inp-ksom[x][y])
    return ksom
```

```python
#generate training dataset using shades of red, green, blues with some shades of yello, teal and pink
train_data=np.array([[255,0,0],
                    [255,99,71],
                    [255,127,80],
                    [205,92,92],
                    [0,100,0],
                    [0,128,0],
                    [34,139,34],
                    [0,255,0],
                    [0,0,255],
                    [65,105,225],
                    [138,43,226],
                    [75,0,130],
                    [255,255,0],
                    [154,205,50],
                    [85,107,47],
                    [107,142,35],
                    [0,128,128],
                    [0,139,139],
                    [0,255,255],
                    [0,255,255],
                    [255,20,147],
                    [255,105,180],
                    [255,182,193],
                    [255,192,203]])
```

```python
#normalize the train data by dividing it by 255
train_data=train_data/255
```

```python
train_data
```

```python
# KSOM_func take sigma value and training data and returns the SOM figures after 0, 20, 40, 100 and 1000 iterations
def KSOM_Func(sig_inp,train_data):   
    ii=0
    ksom_store=np.zeros((5,100,100,3))#used to store som data after given epochs
    ksom=np.random.random((x_len,y_len,feat_size))#randomly initialize the weigths
    ksom_store[ii]=ksom
    for i in range(1,t_epoch+1,1):# initiate the training
        print('Epoch = ',i,'Sigma value = ',sig_inp)
        for sample in range(n_color):# traing every color under one epoch
            dist_mat=np.zeros((x_len,y_len))
            sample_inp=train_data[sample]
            for x in range(x_len):
                for y in range(y_len):
                    current_dist=dist(sample_inp,ksom[x][y])# calculating distance with respect to given color
                    dist_mat[x][y]=current_dist
            indices = np.where(dist_mat == dist_mat.min())# find the winning neuron with its indices
            x_win=indices[0][0]
            y_win=indices[1][0]
            ksom=wt_update(x_win,y_win,i,sig_inp,ksom,sample_inp)#updating the weights
        if i ==20 or i==40 or i==100 or i==1000:
            ii=ii+1
            ksom_store[ii]=ksom #store som data
            
    return ksom_store
```

```python
#Implementing SOM using various sigma values
out_1=KSOM_Func(1,train_data)
out_10=KSOM_Func(10,train_data)
out_30=KSOM_Func(30,train_data)
out_50=KSOM_Func(50,train_data)
out_70=KSOM_Func(70,train_data)
```

```python
#plotting the figures for sigma =1
f = plt.figure(figsize=(15,15))
plt.suptitle('For sigma = 1')
ax1 = f.add_subplot(321)
ax2 = f.add_subplot(322)
ax3 = f.add_subplot(323)
ax4 = f.add_subplot(324)
ax5= f.add_subplot(325)
ax1.imshow(out_1[0])
ax1.title.set_text('Original Grid (Randomly generated)')
ax2.imshow(out_1[1])
ax2.title.set_text('Figure at 20 epochs')
ax3.imshow(out_1[2])
ax3.title.set_text('Figure at 40 epochs')
ax4.imshow(out_1[3])
ax4.title.set_text('Figure at 100 epochs')
ax5.imshow(out_1[4])
ax5.title.set_text('Figure at 1000 epochs')
```

## Figures of SOM with sigma = 1

The above figures show SOM figures at various iterations with sigma value of 1. Here, radius of the neighbourhood is very small so sepration of various color is not seen. No convergence has been found.
